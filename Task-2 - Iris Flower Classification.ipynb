# Separate Input and output columns

```python
data = iris.values

X = data[:,0:4]
y = data[:,4]
```

# Split data into Training and Testing

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
#print(X_test)
```

# Model Selection

## <font color="98AFC7">*Support Vector Machine*</font>

```python
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(X_train,y_train)
```

```python
prediction_svc = model_svc.predict(X_test)

# Calculate the accuracy score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction_svc)*100)
```

## <font color="98AFC7">*Logistic Regression*</font>

```python
from sklearn.linear_model import LogisticRegression

model_logr = LogisticRegression()
model_logr.fit(X_train,y_train)
```

```python
prediction_logr = model_logr.predict(X_test)

# Calculate the accuracy score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction_logr)*100)
```

## <font color="98AFC7">*Decision Tree*</font>

```python
from sklearn.tree import DecisionTreeClassifier

model_dtree = DecisionTreeClassifier()
model_dtree.fit(X_train,y_train)
```

```python
prediction_dtree = model_dtree.predict(X_test)

# Calculate the accuracy score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction_dtree)*100)
```

## <font color="98AFC7">*K Nearest Neighbours*</font>

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier()
model_knn.fit(X_train,y_train)
```

```python
prediction_knn = model_knn.predict(X_test)

# Calculate the accuracy score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction_knn)*100)
```

### <font color="566D7E">*Comparing actual output vs predicted output*</font>

```python
for i in range(len(prediction_knn)):
    print(y_test[i],prediction_knn[i])
```

### <font color="566D7E">*Detailed Classification Report*</font>

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction_knn))
```

### <font color="566D7E">*Confusion Matrix*</font>

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test,y_pred=prediction_svc)
print(conf_matrix)
```

### <font color="566D7E">*Predicting species from input vector*</font>

```python
X_new = np.array([[4.9,2.2,3.8,1.1],[3,2,1,0.2],[5.3,2.5,4.6,1.9]])

prediction = model_logr.predict(X_new)
print("Prediction of Species: {}".format(prediction))
```
